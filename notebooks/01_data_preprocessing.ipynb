```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "../data/raw/network_inventory_sample_data.xlsx"

df_assets = pd.read_excel(file_path, sheet_name="Asset_Inventory")
df_util = pd.read_excel(file_path, sheet_name="Utilization_Records")
df_growth = pd.read_excel(file_path, sheet_name="Asset_Growth_History")


print("Assets shape:", df_assets.shape)
print("Utilization shape:", df_util.shape)
print("Growth shape:", df_growth.shape)
```

```python

print("Missing values in Assets table:")
print(df_assets.isnull().sum())
print("\nMissing values in Utilization table:")
print(df_util.isnull().sum())
print("\nMissing values in Growth table:")
print(df_growth.isnull().sum())

```

```python
print("Duplicate rows in Assets table:", df_assets.duplicated().sum())
print("Duplicate rows in Utilization table:", df_util.duplicated().sum())
print("Duplicate rows in Growth table:", df_growth.duplicated().sum())
```

```python
df_assets["Purchase_Date"] = pd.to_datetime(df_assets["Purchase_Date"])
df_assets["Warranty_Expiry"] = pd.to_datetime(df_assets["Warranty_Expiry"])
df_util["Record_Date"] = pd.to_datetime(df_util["Record_Date"])

print("Assets date columns:")
print(df_assets[["Purchase_Date", "Warranty_Expiry"]].dtypes)
print("\nUtilization date column:")
print(df_util[["Record_Date"]].dtypes)
```

```python
df_assets.to_csv("../data/processed/assets_clean.csv", index=False)
df_util.to_csv("../data/processed/utilization_clean.csv", index=False)
df_growth.to_csv("../data/processed/growth_clean.csv", index=False)

print("All three cleaned files saved to data/processed/ successfully!")
```
