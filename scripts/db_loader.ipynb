```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

password = quote_plus("sql346@")
engine = create_engine(f'mysql+pymysql://root:{password}@localhost:3306/network_inventory')

print("Connected to MySQL successfully!")
```

```python
df_assets = pd.read_csv("../data/processed/assets_clean.csv")
df_util = pd.read_csv("../data/processed/utilization_clean.csv")
df_growth = pd.read_csv("../data/processed/growth_clean.csv")
df_recommendations = pd.read_csv("../data/processed/upgrade_recommendations.csv")
df_forecast = pd.read_csv("../data/processed/forecast_counts.csv")

print("All files loaded successfully!")
print(f"Assets: {len(df_assets)} rows")
print(f"Utilization: {len(df_util)} rows")
print(f"Growth: {len(df_growth)} rows")
print(f"Recommendations: {len(df_recommendations)} rows")
print(f"Forecast: {len(df_forecast)} rows")
```

```python
df_assets.to_sql("asset_inventory", con=engine, if_exists="replace", index=False)
df_util.to_sql("utilization_records", con=engine, if_exists="replace", index=False)
df_growth.to_sql("growth_history", con=engine, if_exists="replace", index=False)
df_recommendations.to_sql("upgrade_recommendations", con=engine, if_exists="replace", index=False)
df_forecast.to_sql("forecast_counts", con=engine, if_exists="replace", index=False)

print("All tables pushed to MySQL successfully!")
```
